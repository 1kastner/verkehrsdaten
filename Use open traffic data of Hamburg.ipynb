```python
url = "http://transport_verkehr.daten-hamburg.de/vkl/vkl.txt"
resp = requests.get(url)
content = resp.text
content
```

So let's parse that data

```python
lines = [line.split(";") for line in content.split("#\r\n") if line]
metadata = lines[:6]
data = lines[6:]

print("Metadata: ", metadata)
print("Data: ", data)
```

```python
translation = {
    "0": "flowing",
    "1": "stop-and-go",
    "2": "heavy",
    "3": "congestion",
    "255": "no data"
}

df = pd.DataFrame(data={
    'location_id': [int(row[0]) for row in data],
    'state_of_traffic': [translation[row[1]] for row in data]
})

df = df.set_index('location_id')
df = df.sort_index()
df
```

Next task is to gather the geopositions of the locations and visualise them.

```python
resp = requests.get("https://geodienste.hamburg.de/HH_WFS_Verkehr_opendata?"
                    + "SERVICE=WFS&REQUEST=GetFeature&VERSION=1.1.0&TYPENAME=bab_vkl")
content = resp.text
print(content)
```

```python
import xml.etree.ElementTree as ET
tree = ET.fromstring(content)
tree
```

```python
data = {
    'location_id': [],
    'coordinates': [],
    'state_of_traffic_2': []
}

namespaces = {"app": "http://www.deegree.org/app", "gml": "http://www.opengis.net/gml"}

for node in tree.findall("./gml:featureMember/app:bab_vkl", namespaces=namespaces):
    vkl_id = int(node.find("app:vkl_id", namespaces=namespaces).text)
    klasse_node = node.find("app:klasse", namespaces=namespaces)
    klasse = klasse_node.text if klasse_node is not None else ""
    klasse_translated = translation.get(klasse, "no data")
    location_node = node.find("./app:the_geom/gml:MultiLineString/gml:lineStringMember/gml:LineString/gml:posList", 
                         namespaces=namespaces)
    location = location_node.text if location_node is not None else ""
    data["location_id"].append(vkl_id)
    data["coordinates"].append(location)
    data["state_of_traffic_2"].append(klasse_translated)
    
df2 = pd.DataFrame(data=data)
df2 = df2.set_index("location_id")
df2 = df2.sort_index()
df2
```

```python
big_df = df.join(df2)
big_df
```
